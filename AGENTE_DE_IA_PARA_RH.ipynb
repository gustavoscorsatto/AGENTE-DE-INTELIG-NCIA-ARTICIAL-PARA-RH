# Pré-processamento

Importando CSV do Google Drive

````python
import pandas as pd

url = 'https://docs.google.com/spreadsheets/d/13wDXB1shiAbQ3txMoiMGIPtjLub-bPjvWw0yQU1Yd48/edit?usp=sharing'

file_id = url.split('/')[-2]

csv_export_url = f'https://docs.google.com/spreadsheets/d/{file_id}/export?format=csv'

df = pd.read_csv(csv_export_url)

print('DataFrame importado corretamente:')
print(df)
````

````python
import pandas as pd
import re
import spacy
from unidecode import unidecode


#df = pd.read_csv('/content/texto_candidatos.csv')


colunas_interesse = df.columns[[6, 7, 8, 34, 35, 44, 45, 46, 47]]


nlp = spacy.load("pt_core_news_sm")

def preprocessamento_text(texto):
    if pd.isna(texto):
        return ""
    texto = unidecode(texto.lower())
    texto = re.sub(r'[^a-zà-ú\s]', ' ', texto)
    doc = nlp(texto)
    tokens = [token.lemma_ for token in doc if not token.is_stop]
    return ' '.join(tokens)

for col in colunas_interesse:
    if df[col].dtype == 'object':
        df[col] = df[col].apply(preprocessamento_text)
    else:
        print(f"Pré-processamento não aplicado para '{col}' pois, tipo: {df[col].dtype}.")

df[colunas_interesse].head(10)
````

# Modelo de Classificação

````python
import os
import getpass
import json
import time
import pandas as pd
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

# --- 1. Configuração da API Groq ---
# Solicita a chave se ela ainda não estiver definida no ambiente
if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Digite sua chave API da Groq: ")

# Inicializa o ChatOpenAI apontando para a Groq
llm = ChatOpenAI(
    openai_api_base="https://api.groq.com/openai/v1", # Endereço oficial da Groq
    openai_api_key=os.environ.get("GROQ_API_KEY"),
    model_name="llama-3.3-70b-versatile", # Modelo Llama 3.3 listado no seu material
    temperature=0.1, # Baixa temperatura para ser mais preciso/determinístico
    max_retries=2
)

print(f"Modelo Ativo na Groq: {llm.model_name}")

# --- 2. Definição do Prompt ---
areas_liga = "RH, Projetos, Marketing, Training"

classification_prompt = PromptTemplate(
    template="""
    Atue como um especialista em Recrutamento e Seleção.
    Analise o texto do candidato abaixo e classifique-o para uma das áreas da Liga: {areas}.

    DADOS DO CANDIDATO:
    {texto_candidato}

    TAREFA:
    1. Identifique a área (Area) mais compatível.
    2. Atribua um Score (0-100) de afinidade.
    3. Dê uma justificativa curta (1 frase).

    RESPOSTA OBRIGATÓRIA EM JSON:
    {{
        "area": "Nome da Área",
        "score": 90,
        "justificativa": "Sua explicação aqui"
    }}
    """,
    input_variables=["areas", "texto_candidato"]
)

chain = classification_prompt | llm

# --- 3. Execução no DataFrame ---
print("Iniciando classificação com Groq (Llama 3)...")

resultados_area = []
resultados_score = []
resultados_justificativa = []

# Supondo que seu df e colunas_interesse já existam do passo de pré-processamento
for index, row in df.iterrows():
    # Cria o perfil juntando as colunas relevantes
    perfil_texto = row[colunas_interesse].apply(str).str.cat(sep=' ')

    try:
        # Chama o modelo
        response = chain.invoke({
            "areas": areas_liga,
            "texto_candidato": perfil_texto
        })

        # Limpeza do JSON (caso o modelo coloque markdown ```json ... ```)
        content = response.content.strip()
        if "```json" in content:
            content = content.split("```json")[1].split("```")[0]
        elif "```" in content:
            content = content.split("```")[1].split("```")[0]

        data = json.loads(content)

        resultados_area.append(data.get('area', 'Indefinido'))
        resultados_score.append(data.get('score', 0))
        resultados_justificativa.append(data.get('justificativa', ''))

        print(f"Candidato {index}: {data.get('area')} (Score: {data.get('score')})")

    except Exception as e:
        print(f"Erro no candidato {index}: {e}")
        resultados_area.append("Erro")
        resultados_score.append(0)
        resultados_justificativa.append("Erro na análise")

    # A Groq é muito rápida, mas respeitamos um limite de segurança
    time.sleep(0.5)

# --- 4. Salvando Resultados ---
df['Classificacao_Sugerida'] = resultados_area
df['Score'] = resultados_score
df['Justificativa'] = resultados_justificativa

# Exibe amostra e salva
print("\nClassificação Concluída!")
print(df[['Nome Completo', 'Classificacao_Sugerida', 'Score']].head())

df.to_csv('classificacao_candidatos_groq.csv', index=False, encoding='utf-8-sig')
print("Arquivo 'classificacao_candidatos_groq.csv' pronto para download.")
````

````python
import os
import json
import time
import pandas as pd
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate

# Configuração da LLM (Llama 3 via Groq)
llm = ChatOpenAI(
    openai_api_base="https://api.groq.com/openai/v1",
    openai_api_key=os.environ.get("GROQ_API_KEY"),
    model_name="llama-3.3-70b-versatile",
    temperature=0.1,
    max_retries=2
)

# --- NOVO PROMPT AVANÇADO ---
classification_prompt = PromptTemplate(
    template="""
    Atue como um Especialista em Recrutamento para uma Liga de Data Science.
    Sua missão é avaliar o perfil do candidato abaixo e atribuir uma nota de 0 a 100 para CADA uma das 4 áreas possíveis, baseando-se nas definições a seguir:

    DEFINIÇÕES DAS ÁREAS:
    1. RH: Gestão de pessoas, empatia, organização, liderança, mediação.
    2. PROJETOS: Habilidade técnica, lógica, Python/SQL, foco em resolver problemas.
    3. MARKETING: Comunicação, criatividade, design, redes sociais, branding.
    4. TRAINING: Didática, ensino, paciência, vontade de compartilhar conhecimento.

    DADOS DO CANDIDATO:
    {texto_candidato}

    TAREFA:
    Analise as respostas e atribua um Score (0-100) para cada área.
    *Nota: Um candidato pode ter notas altas em mais de uma área.*

    RESPOSTA OBRIGATÓRIA EM JSON (SEM MARKDOWN):
    {{
        "score_rh": 0,
        "score_projetos": 0,
        "score_marketing": 0,
        "score_training": 0,
        "area_principal": "Nome da área com maior nota",
        "justificativa": "Explicação breve do perfil"
    }}
    """,
    input_variables=["texto_candidato"]
)

chain = classification_prompt | llm

print("Iniciando avaliação multidimensional...")

# Listas para armazenar os novos dados
scores_rh = []
scores_proj = []
scores_mkt = []
scores_train = []
areas_principais = []
justificativas = []

# Loop de processamento
for index, row in df.iterrows():
    # Criar perfil (ajuste 'colunas_interesse' conforme seu código anterior)
    perfil_texto = row[colunas_interesse].apply(str).str.cat(sep=' ')

    try:
        response = chain.invoke({"texto_candidato": perfil_texto})

        # Limpeza do JSON
        content = response.content.strip()
        if "```json" in content:
            content = content.split("```json")[1].split("```")[0]
        elif "```" in content:
            content = content.split("```")[1].split("```")[0]

        data = json.loads(content)

        # Salvando nas listas
        scores_rh.append(data.get('score_rh', 0))
        scores_proj.append(data.get('score_projetos', 0))
        scores_mkt.append(data.get('score_marketing', 0))
        scores_train.append(data.get('score_training', 0))
        areas_principais.append(data.get('area_principal', 'Indefinido'))
        justificativas.append(data.get('justificativa', ''))

        print(f"Candidato {index}: RH={data.get('score_rh')} | Proj={data.get('score_projetos')} | Mkt={data.get('score_marketing')} | Train={data.get('score_training')}")

    except Exception as e:
        print(f"Erro id {index}: {e}")
        scores_rh.append(0); scores_proj.append(0); scores_mkt.append(0); scores_train.append(0)
        areas_principais.append("Erro"); justificativas.append("Erro")

    time.sleep(0.5)

# Adicionar colunas ao DataFrame
df['Score_RH'] = scores_rh
df['Score_Projetos'] = scores_proj
df['Score_Marketing'] = scores_mkt
df['Score_Training'] = scores_train
df['Area_Sugerida'] = areas_principais
df['Analise_IA'] = justificativas

print("\nAvaliação Concluída!")
````

# Planilha de Resultados

````python
import pandas as pd

# Instalar biblioteca se necessário
# !pip install XlsxWriter

# 1. Garantir que os dados numéricos estão corretos
cols_scores = ['Score_RH', 'Score_Projetos', 'Score_Marketing', 'Score_Training']
for col in cols_scores:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

arquivo_saida = 'Analise_Candidatos_Visual_Melhorado.xlsx'
writer = pd.ExcelWriter(arquivo_saida, engine='xlsxwriter')

# Selecionar e ordenar as colunas
cols_finais = ['Nome Completo'] + cols_scores + ['Area_Sugerida', 'Analise_IA']
df[cols_finais].to_excel(writer, sheet_name='Ranking', index=False)

wb = writer.book
ws = writer.sheets['Ranking']

# --- CRIAÇÃO DE ESTILOS ---

# Estilo do Cabeçalho (Azul Profundo, Fonte Maior)
header_fmt = wb.add_format({
    'bold': True,
    'fg_color': '#203764',
    'font_color': 'white',
    'border': 1,
    'align': 'center',
    'valign': 'vcenter',
    'font_size': 12
})

# Estilo para o Texto (Alinhado ao Topo, Quebra de Linha)
text_fmt = wb.add_format({
    'border': 1,
    'valign': 'top',
    'text_wrap': True,
    'font_size': 11
})

# Estilo Centralizado para Scores e Áreas
center_fmt = wb.add_format({
    'border': 1,
    'align': 'center',
    'valign': 'vcenter',
    'font_size': 11
})

# --- APLICANDO FORMATAÇÃO ---

# 1. Aplicar estilo ao Cabeçalho
for col_num, value in enumerate(cols_finais):
    ws.write(0, col_num, value, header_fmt)

# 2. Ajustar Largura das Colunas (Aumentadas)
ws.set_column('A:A', 40, text_fmt)   # Nome Completo (Mais largo)
ws.set_column('B:E', 18, center_fmt) # Scores (Mais espaçado)
ws.set_column('F:F', 25, center_fmt) # Área Sugerida
ws.set_column('G:G', 80, text_fmt)   # Análise/Justificativa (Bem largo para leitura)

# 3. Mapa de Calor nos Scores (Verde -> Amarelo -> Vermelho)
ws.conditional_format('B2:E1000', {
    'type': '3_color_scale',
    'min_color': '#F8696B', # Vermelho
    'mid_color': '#FFEB84', # Amarelo
    'max_color': '#63BE7B'  # Verde
})

# Salvar
writer.close()
print(f"Planilha '{arquivo_saida}' gerada com visual ampliado!")
````
